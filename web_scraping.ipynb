```python
from bs4 import BeautifulSoup
```

```python
import requests
import pandas as pd
from pandas import Series
```

```python
url = 'https://www.studentnewsdaily.com/archive/daily-news-article/'
```

```python
#step 1 : call the api server using request module
# get req - getting info from server
#post - giving info to server ex input in forms
result = requests.get(url)
```

```python
# print(result.content)
soup = BeautifulSoup(result.content, 'lxml')
```

```python
print(soup)  #prints clean html
```

to fetch some particular section of web, u can fetch by using its/ that div's class name

```python
content = soup.findAll('div')
# print(content)
print(len(content))
```

```python
content = soup.findAll('section')
```
